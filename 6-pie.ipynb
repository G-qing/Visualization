# 一、饼图
## 1.概述：
饼图（Pie Chart）是一种常用的统计图表，用于表示各组成部分占总体的比例。每个扇形的面积表示一个类别的比例，总体的比例加起来等于100%。饼图最早在19世纪由英国统计学家威廉·普雷费尔首次提出，后来被广泛应用于各类数据分析和展示。

* 背景：饼图是通过将一个整体划分成多个扇形区域来表示数据的不同部分，各扇形的角度与对应的数据大小成正比。饼图的主要目的是通过视觉化展示各部分在总体中的占比情况，帮助用户直观地比较不同类别之间的比例差异。

* 用途： 

饼图主要用于：
1. 比例对比：饼图常用于展示各部分在总量中的占比情况，能够清晰地反映出不同部分在整体中的权重。
2. 类别分布：对于少量分类数据，饼图可以有效展示各类别的分布情况，适合显示某一变量的构成。
3. 单一维度数据分析：饼图通常用于单一维度数据分析，尤其是各类占比的可视化。

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

labels = ['公司A', '公司B', '公司C', '公司D']
sizes = [35, 30, 20, 15]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)  

# 创建饼图
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)

plt.title('各公司市场份额比例对比')
plt.axis('equal')  
plt.show()
```

* 说明：
这个饼图显示了四个公司在市场中的份额，通过突出公司A使得该公司更为显眼。饼图使用了百分比格式，可以直接展示比例的大小。

```python
import matplotlib.pyplot as plt

labels = ['男性', '女性']
sizes = [51, 49]
colors = ['#ff9999', '#66b3ff']

# 创建饼图
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=False)

plt.title('学生性别构成')
plt.axis('equal')  
plt.show()
```

* 说明：
此饼图展示了一个学生群体的性别构成，分别是男性和女性的比例。通过这种方式，可以直观地看到男女生的比例分布。

* 适用场景
    1. 简单的分类数据：当数据类别较少（通常不超过5-7个）时，饼图是一个良好的选择。若类别过多，饼图的可读性会大幅下降。
    2. 对比类别的相对大小：适合用来展示类别间相对大小的差异。例如，市场份额分析、人口构成、收入来源构成等。
    3. 直观展示比例关系：在想快速向观众展示某个部分在整体中的占比时，饼图是很好的选择。

* 不适用的场景
    1. 数据量较大：饼图不适合展示超过7个类别的数据，因为过多的扇形会使图形变得复杂，难以辨别。
    2. 显示精确的数值比较：如果需要精确地比较各部分之间的差异，条形图或柱状图可能更为合适，因为它们能够更直观地展示数值大小的差异。
    3. 负值和零值数据：饼图无法显示负值或零值，因此不适用于此类数据的可视化。

## 2.示例

```python
import pandas as pd
df = pd.read_csv("Electronic_sales_Sep2023-Sep2024.csv")
df
```

* 我们以一组客户购买电子设备的行为数据为例，用支付方式（Payment Method）一列，通过饼图来展示各种支付方式的分布情况。

```python
import matplotlib.pyplot as plt

df['Payment Method'] = df['Payment Method'].str.replace('Paypal', 'PayPal')

payment_method_counts = df['Payment Method'].value_counts()  # 统计各支付方式的频次
labels = payment_method_counts.index  
sizes = payment_method_counts.values  
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  

# 创建饼图
plt.figure(figsize=(6, 6))
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

plt.legend(wedges, labels, title="支付方式", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('支付方式分布')
plt.axis('equal')  
plt.show()
```

上图中可以看出，使用信用卡支付（Credit Card）的付款方式占比最高，为29.3%，其次是和信用卡支付方式相当的支付宝方式（PayPal），占29%。最少使用的支付方式为使用扣账卡（Debit Card），占12.4%，与使用现金（Cash）的付款方式占比相当，使用现金的占比为12.5%。

* 我们以忠诚会员（Loyalty Member）一列，通过饼图来展示会员的分布情况。

```python
import matplotlib.pyplot as plt

loyalty_counts = df['Loyalty Member'].value_counts()
labels = loyalty_counts.index
sizes = loyalty_counts.values
colors = ['#ff9999', '#66b3ff']  

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=[0.1 if label == 'Yes' else 0 for label in labels])

plt.title('忠诚会员状态分布')
plt.legend(title="忠诚会员", loc="upper left")
plt.axis('equal')
plt.show()
```

可以看到会员占大约1/5.

## 3.绘制方法
下面是使用不同库绘制基本饼图的示例，包括绘图方法和函数参数的详细介绍。
1. 使用matplotlib


matplotlib 是最常用的绘图库之一，使用 plt.pie() 函数绘制饼图。

```python
import matplotlib.pyplot as plt

# 示例数据
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = [0.1, 0, 0, 0]  # 将第一个扇区突出显示

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90, shadow=True)

plt.title('Matplotlib饼图示例')

# 确保饼图是圆形
plt.axis('equal')

# 显示图例
plt.legend(loc="upper right")

plt.show()
```

主要参数说明：
* x: 数值数据，表示每个扇区的大小。
* labels: 每个扇区的标签。
* colors: 扇区的颜色，可以是颜色代码列表。
* explode: 控制每个扇区的突出程度，可以通过设置值为 0 或非零值来控制突出显示。
* autopct: 控制每个扇区中显示的数值格式，'%1.1f%%' 表示显示1位小数的百分比。
* startangle: 起始角度，控制饼图的旋转。
* shadow: 是否在饼图下方显示阴影，True 或 False。
* radius: 控制饼图的半径，默认值是 1。


更多参数：
* counterclock: 布尔值，指定扇区是否按逆时针方向绘制，默认 True。
* pctdistance: 控制标签百分比离饼图中心的距离，默认为 0.6。
* labeldistance: 控制标签离饼图中心的距离。
* normalize: 设置为 True 时会归一化所有值，使其总和为 1。

2. 使用Plotly


plotly 是一个功能强大的交互式绘图库，适合用于 Web 应用。

```python
import plotly.graph_objects as go

# 示例数据
labels = ['A', 'B', 'C', 'D']
values = [30, 20, 25, 25]

# 绘制饼图
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.3)])

fig.update_layout(title_text='Plotly饼图示例')
fig.show()
```

主要参数说明：
* labels: 饼图各扇区的标签。
* values: 对应扇区的值，定义每个扇区的大小。
* hole: 饼图的中心孔大小，设置为 0.3 时变为环形图。
* marker_colors: 控制扇区的颜色，可以是颜色代码的列表。
* pull: 类似于 explode 参数，用来将某些扇区拉出。


更多参数：
* direction: 控制绘图方向，默认为 'counterclockwise'。
* sort: 布尔值，是否按值的大小排序，默认为 True。
* hoverinfo: 设置鼠标悬停时显示的信息，默认显示 label+percent+value。
* textinfo: 控制图中显示的文本，值包括 'label'、'value'、'percent' 等。
* textposition: 设置文本标签的位置，默认 'inside'，还可以设置 'outside'。

3. 使用Altair


Altair 是一个基于声明式设计的库，使用起来非常简洁。

```python
import altair as alt
import pandas as pd

# 示例数据
data = pd.DataFrame({
    'Type': ['A', 'B', 'C', 'D'],
    'Values': [30, 20, 25, 25]
})

# 绘制饼图
chart = alt.Chart(data).mark_arc().encode(
    theta=alt.Theta(field='Values', type='quantitative'),
    color=alt.Color(field='Type', type='nominal'),
    tooltip=['Type', 'Values']
).properties(
    title='Altair饼图示例'
)

# 显示图表
chart.show()
```

主要参数说明：
* theta: 设置每个扇区的角度，通常为定量数据。
* color: 控制每个扇区的颜色，通常使用类别数据。
* tooltip: 设置鼠标悬停时显示的提示信息。


更多参数：
* stroke: 设置扇区边界的颜色。
* fillOpacity: 设置扇区颜色的透明度。

## 4.总结
不同的可视化库提供了多种绘制饼图的方式，matplotlib 提供基础绘图功能，plotly 提供交互式绘图，而 altair 则提供声明式图形生成。这些库都具有灵活的参数设置，可以根据具体的需求进行调整。
