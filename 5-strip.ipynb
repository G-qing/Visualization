# 一、带状图
## 1.概述：
带状图（又称为一维散点图），是针对一维数据的散点图，它本质是数据与固定值（固定x或固定y）之间的散点图，这样形成的图形外观是带状的，因此称为带状图。它是一种用于数据可视化的图表，主要用于展示数据的分布和模式。以下是关于带状图的背景、用途及适用场景的介绍：


* 背景：带状图主要用于显示数值数据的分布情况，尤其是在需要展示单一变量时。它通过在一条线上绘制点，能够直观地展示数据点在该维度上的分布。这种图表形式可以追溯到早期的统计图表，但在现代数据分析中，带状图被广泛用于揭示数据的模式和趋势。

* 用途： 

带状图主要用于：
1. 数据分布: 带状图能够有效地展示数据的集中趋势、离散程度和分布形状。
2. 异常值检测: 通过观察数据点的分布，可以快速识别出潜在的异常值或离群点。
3. 类别比较: 在分组情况下，带状图可以用来比较不同类别的数值分布情况。
4. 趋势分析: 带状图可以帮助分析某一变量随时间或其他变量的变化趋势。

例1：显示数据分布

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

data = np.random.normal(loc=0, scale=1, size=100)

plt.figure(figsize=(10, 6))
plt.plot(data, np.zeros_like(data), 'o', alpha=0.5)
plt.title('数据分布示例')
plt.xlabel('数值')
plt.ylabel('频率')
plt.yticks([]) 
plt.show()
```

例2: 比较不同类别的数据分布

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=1, scale=1.5, size=100)

plt.figure(figsize=(10, 6))
plt.plot(data1, np.zeros_like(data1) + 0.1, 'o', alpha=0.5, label='类别 A')
plt.plot(data2, np.zeros_like(data2) - 0.1, 'o', alpha=0.5, label='类别 B')
plt.title('不同类别数据分布比较')
plt.xlabel('数值')
plt.ylabel('类别')
plt.yticks([-0.1, 0.1], ['类别 B', '类别 A']) 
plt.legend()
plt.show()
```

这两个例子分别展示了带状图的两种主要用途：第一个示例直观地展示数据的分布情况，第二个示例比较了两个不同类别的数据分布。

* 适用场景
1. 单变量分析: 带状图适用于展示单一变量的分布，例如学生成绩、收入水平等。
2. 定量数据比较: 当需要比较不同组或类别的数据时，可以使用带状图展示各组数据的分布情况。
3. 探索性数据分析: 在数据分析的初步阶段，带状图可以帮助分析师快速识别数据的基本特征和趋势。
4. 生物统计学: 在医学或生物统计学中，带状图可以用于展示生物测量数据的分布，如体重、血压等。
5. 社会科学研究: 在社会科学研究中，带状图可以帮助研究人员展示与人口统计相关的数值数据分布。

## 2.示例

```python
import pandas as pd
df = pd.read_csv("StudentPerformanceFactors.csv")
df
```

我们以一组学生表现因素数据为例，用父母参与度（Parental_Involvement）与考试分数（Exam_Score）两列，通过带状图与箱线图来展示学生考试分数的整体分布情况。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

for involvement in df['Parental_Involvement'].unique():
    sns.stripplot(x='Parental_Involvement', y='Exam_Score', 
                  data=df[df['Parental_Involvement'] == involvement],
                  jitter=True, alpha=0.5, label=involvement)

sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=df, 
            color='white', fliersize=0, linewidth=1.5)

plt.title('图1 一维散点图与箱线图的结合')
plt.xlabel('父母参与度')
plt.ylabel('考试分数')
plt.legend(title='父母参与度')
plt.show()
```

```python
df['Jittered_X'] = df['Parental_Involvement'].astype('category').cat.codes + np.random.uniform(-0.2, 0.2, size=len(df))

# 创建图形
plt.figure(figsize=(12, 6))

# 绘制随机打乱的带状图（散点图）
sns.stripplot(x='Jittered_X', y='Exam_Score', data=df, 
              jitter=False, alpha=0.5, color='b')

    
plt.xticks(ticks=np.arange(len(df['Parental_Involvement'].unique())),
           labels=df['Parental_Involvement'].unique())

plt.title('图2 随机打乱的带状图')
plt.xlabel('父母参与度')
plt.ylabel('考试分数')
plt.show()
```

我们再次使用了学生表现因素数据，图1 展示了分类的带状图，并且在图中放置了箱线图作为对比。如果只观察箱线图，无法得知数据在若干位置有重复，只知道数据四分位点的位置，而带状图则可以告诉我们哪些位置分别有多少数据点。图2 为随机打乱的带状图，作图方法只是给x轴方向上的固定数值添加了随机数，使原本重叠在一起的数据重新拥有不同的横坐标，从而将重叠的数据分开。

## 3.绘制方法
带状图通常是使用 seaborn 库中的 stripplot() 函数绘制的，但我们也可以使用 matplotlib 和 plotly 等其他库来实现。下面是不同库绘制带状图的基本方法以及相关参数的介绍。
1. 使用 Seaborn 绘制带状图


seaborn 是一个基于 matplotlib 的高级可视化库，提供了简化的绘图接口。

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 绘制带状图
plt.figure(figsize=(10, 6))
sns.stripplot(x='Parental_Involvement', y='Exam_Score', data=df, 
               jitter=True, alpha=0.7, color='blue', size=8)
plt.title('带状图示例 - Seaborn')
plt.xlabel('父母参与度')
plt.ylabel('考试分数')
plt.show()
```

主要参数：
* x: 类别变量，表示 x 轴数据。
* y: 数值变量，表示 y 轴数据。
* data: 数据框，提供数据。
* jitter: 布尔值或浮点数，表示是否添加随机抖动（可以使点分开）。如果为浮点数，则表示抖动范围。
* alpha: 透明度（0 到 1），设置点的透明度。
* color: 点的颜色。
* size: 点的大小。
* dodge: 布尔值，表示是否分开不同类别的点。

2. 使用 Matplotlib 绘制带状图


matplotlib 是 Python 的基本绘图库，尽管不专门为带状图设计，但可以使用 scatter() 方法来实现。

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 为x轴添加抖动
df['Jittered_X'] = df['Parental_Involvement'].astype('category').cat.codes + np.random.uniform(-0.1, 0.1, size=len(df))

# 绘制带状图
plt.figure(figsize=(10, 6))
plt.scatter(df['Jittered_X'], df['Exam_Score'], alpha=0.7, color='blue', s=100)
plt.xticks(ticks=np.arange(len(df['Parental_Involvement'].unique())),
           labels=df['Parental_Involvement'].unique())
plt.title('带状图示例 - Matplotlib')
plt.xlabel('父母参与度')
plt.ylabel('考试分数')
plt.show()
```

主要参数:
* x: 类别变量的数值编码。
* y: 数值变量，表示 y 轴数据。
* alpha: 透明度（0 到 1）。
* color: 点的颜色。
* s: 点的大小（以像素为单位）。
* marker: 点的标记样式。

3. 使用 Plotly 绘制带状图


plotly 是一个用于创建交互式图表的库，可以生成丰富的视觉效果。

```python
import pandas as pd
import plotly.express as px
import numpy as np

df['Jittered_X'] = df['Parental_Involvement'].astype('category').cat.codes + np.random.uniform(-0.1, 0.1, size=len(df))

# 绘制带状图
fig = px.strip(df, x='Jittered_X', y='Exam_Score', color='Parental_Involvement',
                title='带状图示例 - Plotly')
fig.update_traces(marker=dict(size=10, opacity=0.7))
fig.update_layout(xaxis=dict(tickmode='array', tickvals=np.arange(len(df['Parental_Involvement'].unique())),
                              ticktext=df['Parental_Involvement'].unique()))

fig.show()
```

主要参数:
* data_frame: 数据框，提供数据。
* x: 类别变量。
* y: 数值变量。
* color: 分类变量，用于区分不同类别的点。
* opacity: 透明度（0 到 1）。
* size: 点的大小。
* jitter: 布尔值，表示是否添加随机抖动。

## 4.总结
这三个库提供了不同的方法来绘制带状图。以下是各个库的特点和参数介绍：


* Seaborn: 适合统计数据可视化，参数丰富，支持高层次的可视化。
* Matplotlib: 适合定制化的可视化，基础绘图库，灵活性高。
* Plotly: 适合创建交互式图表，用户体验好，支持缩放和平移。
