# 一、风玫瑰图
## 1.概述：
风玫瑰图（Wind Rose）是一种用于展示风速和风向分布的图表，常用于气象学和环境科学领域。它是一种极坐标图，通常在中心点代表无风，径向方向表示风向，而半径的大小或柱状的长度表示该风向下的风速或频率。不同的颜色通常用于区分不同风速范围，使得数据的可视化更清晰。


风玫瑰图的本质是堆叠的直方图，只是绘制在极坐标系中。它通常把水平方向以0°到360°计，并平分为16个风向区间，用16个扇形（或指针或数据点）表示来自这些区间的风，在每一扇形上对各等风速进行频数汇总，最后除以观测时间内的观测次数（包括无风次数），就得到了各个风向上各种风速的频率，用扇环的半径表示出来即可。

* 背景与用途：
1. 气象分析：风玫瑰图最常用于气象学，帮助描述一个地区的风力模式。通过展示不同方位上的风速和频率，它能够清晰反映出某个区域的常见风向以及强风出现的概率。
2. 风力发电：风电行业广泛使用风玫瑰图来分析风资源分布，从而确定风力发电场的最佳选址。通过风玫瑰图，工程师可以了解到某个区域的风能利用潜力，决定涡轮机的摆放方向及类型。
3. 空气污染研究：在环境科学中，风玫瑰图用于分析空气污染物的扩散途径，帮助追踪污染源。风向与风速直接影响空气中的颗粒物如何传播，因此风玫瑰图能为污染防控提供有力数据支撑。
4. 建筑与城市规划：建筑师和城市规划师会利用风玫瑰图优化建筑设计，合理选择建筑朝向，确保通风效果并减少能源消耗。

* 适用场景
1. 长期风向与风速趋势分析：例如，评估某地一年四季的风速和风向分布情况。
2. 风资源评估：帮助风电场规划和建设，找到风能资源集中的区域。
3. 污染控制策略：在工业和城市区域，追踪风速与方向的变化，帮助设计污染控制措施。
4. 交通与物流：分析港口、机场等交通枢纽的风向，优化船舶、飞机的调度和航行安全。

## 2.示例

```python
import pandas as pd
df = pd.read_csv("weatherHistory.csv")
df
```

上表是 Szeged 2006-2016年的天气数据。这里选用风速和风向两列。

```python
df['Wind Speed (km/h)'] = df['Wind Speed (km/h)'].round(2)
data = df[['Wind Speed (km/h)','Wind Bearing (degrees)']]
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes
import numpy as np

wind_directions = df['Wind Bearing (degrees)']
wind_speeds = df['Wind Speed (km/h)']

ax = WindroseAxes.from_ax()

# 设置风速分区为0-5, 5-10, 10-15, 15-20, 20-25 km/h，风向分为16个扇区
ax.bar(wind_directions, wind_speeds, normed=True, opening=0.8, edgecolor='white', bins=np.arange(0, 25, 5), nsector=16)

cmap = plt.get_cmap('viridis')  
ax.set_legend(title="Wind Speed (km/h)")
plt.title('Wind Rose (16 Wind Directions) with Wind Speed Ranges')

plt.show()
```

上图为当地风玫瑰图示例，各种风速的频率分别用一个扇环来表示。也就是说，某个扇环的半径越长，表示该风速水平上的刮风次数越多。从上图中可以看出，该地区最常刮的是东南偏南风，因为该方向的扇环半径最长；相对来说东北偏东风较少。无风的概率大概是0.2%。

风速的范围从 0 到大于20 km/h 不等，图中不同的颜色代表不同的风速区间。


较高风速（黄色表示，> 20 km/h）集中在 西北方向（N-W）和东南方向（S-E），显示这些方向出现强风的频率较高。


其他方向的风速分布较为平均，大多数情况下，风速保持在 5-15 km/h 之间。

风向主要集中在 西北（N-W）、东南（S-E）和东北（N-E），这意味着这些方向上的风频率较高。东南方向的风速区间相对较广，风速从较低值逐渐增加，显示出持续性较强的风。
东北和西北方向的风速范围比较集中，这表明这些方向上的风速有较大的波动。

也就是说，如果要在这座城市附近修建会造成空气污染的化工厂，那么应该避免在西北和东南下风处设立居民区或环保设施，因这些区域风速高，可能会快速传播污染物。


可以将污染源布置在西南或东北，风速较低，有助于控制污染物扩散。

另外，如果要建设风力发电厂，因其通常选址于风速较高且风向较为稳定的区域。从图中看，西北（N-W）和东南（S-E） 的风速较高，适合布局风力发电设备。


特别是东南方向，风速从较低到高的波动范围较大，适合设置不同容量的风力发电机组。

## 3.绘制方法
通常使用 Python 中一个专门的库叫做 windrose，它非常方便用于绘制风玫瑰图。windrose 库基于 matplotlib，使得创建风玫瑰图变得更加简便，尤其适用于处理气象数据。

* windrose 库的特点：
1. 专注于风玫瑰图：它封装了所有生成风玫瑰图的细节，简化了复杂的极坐标绘图工作。
2. 多种可视化选项：除了基本的风玫瑰图外，windrose 还支持多种样式的图形展示，如柱状、轮廓线和填充风玫瑰图，能更灵活地表示不同风速和风向数据。
3. 与matplotlib兼容：由于它基于matplotlib，因此可以和其他图形元素无缝集成，例如添加标题、标注等。
4. 支持不同格式的数据输入：可以方便地处理来自数据框、列表或数组的风速和风向数据。

windrose 库提供了一些方便的函数来绘制和定制风玫瑰图。以下是几个常用的函数及其用法说明：

1. WindroseAxes.from_ax()


这个函数用于创建一个风玫瑰图的坐标轴，返回一个 WindroseAxes 对象，你可以在该对象上进行风玫瑰图的绘制操作。

```python
from windrose import WindroseAxes

ax = WindroseAxes.from_ax()  # 创建一个风玫瑰图的坐标轴
```

你可以把它看作是一个包含极坐标的绘图区域，后续的图形绘制都将在这个坐标轴上进行。

2. bar(direction, var, **kwargs)


bar() 函数是绘制柱状风玫瑰图的核心方法，它会根据提供的风向和风速数据生成图形。这个方法最常见，用于直观展示每个方向上的风速分布。


参数：
* direction：风向数据（通常是 0 到 360 度）。
* var：对应的风速数据。
* normed：布尔值，是否对数据进行标准化，使得柱的高度表示相对频率（默认为 True）。
* opening：控制每个扇区的开口宽度，取值范围为 0 到 1，默认值为 0.8。
* edgecolor：柱状图的边缘颜色。

```python
import numpy as np
from windrose import WindroseAxes

wind_speed = np.random.random(500) * 20  # 风速数据
wind_direction = np.random.random(500) * 360  # 风向数据

ax = WindroseAxes.from_ax()  # 创建风玫瑰图坐标轴
ax.bar(wind_direction, wind_speed, normed=True, opening=0.8, edgecolor='white')  # 绘制风玫瑰图
ax.set_legend()  # 添加图例
plt.show()
```

3. set_legend()


set_legend() 函数用于设置图例。它自动生成风速的图例，展示各个风速区间的颜色对应关系。


参数：
* loc：设置图例的位置（例如 'best', 'upper left', 'lower right'）。
* title：设置图例标题，通常用于描述风速的单位。

```python
ax.set_legend(title="风速 (m/s)", loc="lower left")  # 设置图例
```

4. set_yticks([])


set_yticks([]) 用于设置径向方向的刻度标签（表示频率），通常用于调整图形的刻度或隐藏刻度。

```python
ax.set_yticks([])  # 隐藏径向刻度
```

## 四、总结：
* WindroseAxes.from_ax()：创建风玫瑰图的坐标轴。
* bar()：绘制柱状风玫瑰图。
* set_legend()：设置图例。
* set_yticks()：设置或隐藏径向刻度。
这些函数使得 windrose 库在绘制风速、风向数据的可视化方面非常强大和灵活。
