# 一、小提琴图
## 1.概述：
小提琴图（violin plot）是一种结合了箱线图（box plot）和核密度估计（kernel density estimate, KDE）的图形，用于显示数据分布及其概率密度。它通过图形展示数据分布的形状、中心趋势以及数据的扩展情况，特别适合对比多个组之间的数据分布。


* 背景：小提琴图的名字来源于其外观，因为它的外观有时候与小提琴比较像（尤其在展示双峰数据的密度时）。最早在统计学领域应用于显示数据分布。与传统的箱线图不同，小提琴图在展示数据集中信息（如四分位数、极值）的同时，还通过对称的概率密度曲线揭示数据的分布趋势。

* 用途： 

小提琴图主要用于：
1. 数据分布的可视化：通过展示密度曲线，直观地揭示数据的分布是否是对称的、是否有多模态（多个峰值）等。
2. 数据集之间的对比：可以直观比较不同组的数据分布，特别是在多个数据集或分组情况下，展示其均值、中位数和数据的分布情况。
3. 处理较大样本的分布：相比箱线图，小提琴图能够更好地处理大数据集，且对数据中存在的离群点较为鲁棒。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 6))
sns.violinplot(data=data)
plt.title("图1 数据分布的可视化", fontsize=16)
plt.xlabel("数据分布", fontsize=14)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

group1 = np.random.normal(loc=0, scale=1, size=1000)
group2 = np.random.normal(loc=1, scale=1.5, size=1000)
group3 = np.random.normal(loc=-1, scale=0.5, size=1000)

df = pd.DataFrame({
    '值': np.concatenate([group1, group2, group3]),
    '组别': ['组1'] * 1000 + ['组2'] * 1000 + ['组3'] * 1000
})

plt.figure(figsize=(8, 6))
sns.violinplot(x='组别', y='值', data=df, palette=["yellow", "blue", "green"])
plt.title("图2 数据集之间的对比", fontsize=16)
plt.xlabel("组别", fontsize=14)
plt.ylabel("值", fontsize=14)
plt.show()
```

以上示例中 


例子1的代码生成了一个正态分布的数据集，并使用小提琴图展示其数据的分布情况，如图1，可以清楚看到数据的密度分布。


例子2的代码生成了三个不同分组的正态分布数据集，并通过小提琴图展示它们的分布差异，如图2，能够直观地比较不同组的数据形态。

* 适用场景
    1. 多个分组的比较：当需要对比多个数据组的分布形态时，小提琴图比单纯的箱线图提供了更多信息。
    2. 数据分布复杂的情况：如果数据呈现多个峰值或是不规则分布（非正态分布），小提琴图能够清晰展现这些信息。
    3. 非对称分布或多峰数据集：它适用于揭示非对称分布和多峰数据的复杂特征，而这些特征通常难以从传统的箱线图中观察到。

## 2.示例

```python
import pandas as pd
df = pd.read_csv("StudentPerformanceFactors.csv")
df
```

我们以一组学生表现因素数据为例，用考试分数（Exam_Score）一列，通过小提琴图来展示学生考试分数的整体分布情况。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 6))
sns.violinplot(y='Exam_Score', data=df, color="#3498db")  
plt.title("学生考试分数的整体分布", fontsize=16)
plt.ylabel("考试分数", fontsize=14)
plt.show()
```

从上图可以看出，学生考试成绩基本分布于65-70分这个区间。

用教师质量（Teacher_Quality）和考试分数（Exam_Score）两列，通过小提琴图来展示不同教师质量下的考试成绩分布。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

teacher_quality_order = ['Low', 'Medium', 'High']
df['Teacher_Quality'] = pd.Categorical(df['Teacher_Quality'], categories=teacher_quality_order, ordered=True)

plt.figure(figsize=(10, 6))
sns.violinplot(x='Teacher_Quality', y='Exam_Score', data=df, palette="Set2")
plt.title("不同教师质量下的考试成绩分布", fontsize=16)
plt.xlabel("教师质量", fontsize=14)
plt.ylabel("考试分数", fontsize=14)
plt.show()
```

上图展示了不同教师质量下学生考试成绩的分布情况：


每个教师质量类别对应的部分展示了考试成绩的分布情况。小提琴图的宽度表示该分数段的密度，宽度越大，表示在该分数范围内的学生数量越多。


从图中可以看到，所有三个类别的分数范围大致在60到100之间，低质量教师下的学生考试分数相对集中在较低的区域，而中等和高质量教师下的分数则表现出较大的分散性。


通过比较不同教师质量下的考试成绩分布，可以观察到趋势：通常情况下，教师质量越高，学生的考试成绩中位数也越高，表明教师质量对学生成绩的影响可能是正向的。

## 3.绘制方法
下面是使用不同库绘制基本小提琴图的示例，包括绘图方法和函数参数的详细介绍。
1. 使用matplotlib


虽然 Matplotlib 本身不直接支持小提琴图的绘制，但可以通过其扩展库 seaborn 来实现。下面是使用 Matplotlib 的方法：

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# 绘制小提琴图
plt.figure(figsize=(10, 6))
parts = plt.violinplot(data, showmeans=False, showmedians=True)

# 设置图表标题和标签
plt.title("小提琴图示例", fontsize=16)
plt.xlabel("类别", fontsize=14)
plt.ylabel("值", fontsize=14)
plt.xticks([1, 2, 3], ['类别 1', '类别 2', '类别 3']) 

# 添加图例
plt.legend([parts['bodies'][0]], ['小提琴图'])
plt.show()
```

主要参数说明：
* showmeans：布尔值，设置为 True 时显示均值点，默认为 False。
* showmedians：布尔值，设置为 True 时显示中位数线，默认为 False。
* bw_method：控制小提琴图的平滑程度，可以是一个浮点数或者字符串（如 'scott' 或 'silverman'）。
* positions：指定小提琴图在 x 轴上的位置，可以为列表。

2. 使用Seaborn 

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv("StudentPerformanceFactors.csv")

# 绘制小提琴图
plt.figure(figsize=(10, 6))
sns.violinplot(x='Teacher_Quality', y='Exam_Score', data=df, palette="Set2", inner="quartile")
plt.title("不同教师质量下的考试成绩分布", fontsize=16)
plt.xlabel("教师质量", fontsize=14)
plt.ylabel("考试分数", fontsize=14)
plt.show()
```

主要参数说明：
* x：指定 x 轴的分类变量（此处为教师质量）。
* y：指定 y 轴的连续变量（此处为考试分数）。
* data：传入的数据框（DataFrame）。
* palette：设置颜色调色板，可以选择不同的调色方案（如 "Set2", "Blues", "pastel" 等）。
* inner：控制内部显示，设置为 "quartile" 显示四分位数，设置为 "point" 显示每个数据点。

3. 使用 Plotly 绘制小提琴图

```python
import plotly.express as px

fig = px.violin(df, x='Teacher_Quality', y='Exam_Score', color='Teacher_Quality', 
                 box=True, points="all", title="不同教师质量下的考试成绩小提琴图")
fig.show()
```

主要参数说明：
* x：指定 x 轴的分类变量（此处为教师质量）。
* y：指定 y 轴的连续变量（此处为考试分数）。
* color：根据教师质量为小提琴图上色。
* box：布尔值，设置为 True 显示箱体。
* points：控制显示的数据点，设置为 "all" 显示所有点，设置为 "outliers" 显示异常值。

## 4.总结
以上代码展示了使用不同库（Matplotlib、Seaborn 和 Plotly）绘制小提琴图的方式。在不同的库中，绘制函数和参数的选择会有所不同：
* Matplotlib：主要用于基本的绘图，灵活性高，适合自定义细节，但小提琴图的实现较为复杂。
* Seaborn：基于 Matplotlib，提供美观的默认样式和简便的接口，适合统计绘图。
* Plotly：提供交互式图形，适合展示复杂数据，参数使用灵活方便。


根据需求选择合适的库进行绘图，可以通过不同参数的使用来增强图表的表达效果。
