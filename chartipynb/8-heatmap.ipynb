# 一、聚类分析与热图
## 1.热图背景：
热图是一种二维图像，利用颜色强度表示数值大小或频率。颜色通常会从低到高依次变化，来突出不同区域的值的差异。这种直观的方式帮助用户快速识别数据中的趋势、聚集点和异常值。

* 用途：


热图广泛用于展示数据矩阵、相关性矩阵或多维数据。通过颜色深浅变化，便于观察不同维度间的相互关系。典型的应用场景包括：
1. 生物信息学：基因表达数据展示不同基因在各样本中的表达水平。
2. 金融市场分析：展示股市或外汇的波动情况。
3. 社会科学：人口密度、社交媒体活跃度等。


等等

* 适用场景


适用于需要快速识别数据趋势、局部高值、低值或分布模式的场合，尤其是在有大量数据的情况下。

## 2. 聚类分析：
* 目的：聚类分析的目标是将数据分成若干簇，使同一簇内的数据相似度较高，而不同簇间的相似度较低。这种方法能帮助揭示数据的潜在分布结构。

* 作用：
聚类分析在多个领域有广泛作用，包括：

    1. 图像处理：如图像分割，聚类可用于区分不同的图像区域。
    2. 市场细分：帮助识别消费群体，为不同群体设计专门的市场策略。
    3. 推荐系统：根据用户特征，将其归入某一类用户群，提供个性化推荐。

* 用法：常用的聚类算法有K均值、层次聚类、密度聚类等。一般步骤包括数据预处理、选择合适的聚类算法、调整参数获得最佳聚类结果。聚类分析的结果往往结合热图可视化，便于理解数据的内在模式和结构。

下面是使用Python和常用库（如scikit-learn和matplotlib）进行聚类分析的代码示例。这里我们使用K均值（K-Means）聚类来展示聚类分析的具体用法和结果可视化。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 1. 生成示例数据
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. 可视化原始数据
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("原始数据分布")
plt.show()

# 3. 使用KMeans算法
kmeans = KMeans(n_clusters=4)  # 设定要分的簇数
kmeans.fit(X)  # 拟合模型
y_kmeans = kmeans.predict(X)  # 预测每个样本的簇标签

# 4. 可视化聚类结果
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')  # 颜色显示不同的簇
centers = kmeans.cluster_centers_  # 获取每个簇的中心点
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # 用红色标记簇中心
plt.title("K-Means聚类结果")
plt.show()
```

代码说明


1. 生成数据：使用make_blobs生成一个具有四个簇的二维数据集。
2. 原始数据可视化：绘制散点图来展示数据的初始分布。
3. 模型拟合和预测：
    * 用KMeans对象指定簇的数量。
    * fit方法对数据进行聚类拟合，predict方法生成每个数据点的簇标签。
4. 结果可视化：
    * 用不同颜色表示每个簇。
    * 在图中添加每个簇的中心点，以帮助观察聚类效果。

    
解释


* K-Means算法将数据分成指定的4个簇，并在结果图中展示各簇的中心点。
* 这种分析帮助我们更好地理解数据的分布结构，并在实际应用中用于分类、分组等任务。

# 3. 示例

以下是汽车相关数据，包含各车型的各种参数

```python
import pandas as pd

df = pd.read_csv("Automobile.csv",index_col=None)
df = df.head(30)
df
```

```python
print(df.columns)
```

热图的基础是颜色图，但它在颜色图上做了一个特殊处理—————聚类。具体来说，热图也是用颜色表达矩阵中的单元格数值，如颜色越深表示数值越大，但热图并非只是简单地表达数值大小，而是对矩阵的行或列进行层次聚类，获得聚类的结果之后将行或列以聚类的顺序排列，并在颜色图的边界区域加上聚类的谱系图。

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage
from sklearn.preprocessing import StandardScaler

df.set_index('name', inplace=True)
numeric_df = df.select_dtypes(include=['float64', 'int64']).copy()

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# 层次聚类
# 对行（汽车品牌）进行聚类
row_linkage = linkage(scaled_data, method='ward')
# 对列（参数）进行聚类
col_linkage = linkage(scaled_data.T, method='ward')

# 绘制聚类热图，添加行和列系谱图
sns.clustermap(scaled_data, row_linkage=row_linkage, col_linkage=col_linkage, 
               cmap="viridis", figsize=(12, 10), xticklabels=numeric_df.columns, yticklabels=df.index, 
               dendrogram_ratio=(.2, .2))

# 显示图形
plt.show()
```

该热力图及其聚类结果反映了汽车数据集中不同参数在各个车型之间的差异，具体可以从以下几个方面分析：


1. 热力图的颜色含义：
* 热力图中的颜色代表标准化后的数值大小（例如，油耗、重量、马力等特征的相对值）。
颜色越浅（接近黄色），代表该特征在该车款中的值较高；颜色越深（接近紫色），则表示该特征的值较低。


* 通过颜色变化可以快速看出一些车型在某些特征上的数值是否异常或显著高于其他车型，例如，黄色区域代表某些车款在该特征上数值偏高。

2. 聚类结果的含义：
热力图左侧和顶部的树状图展示了层次聚类的结果，这些分支结构代表了相似车辆和相似特征的聚合。
* 行聚类（车型的聚类）：
行聚类根据各车型在所有数值特征上的相似性来分组。相似的车型在特征上更接近，因此会被分到同一组。可以观察到相似车型的聚集，显示出不同车型在动力（如马力和加速度）、重量和油耗等方面的区别。
例如，如果某些小型、轻便的车辆聚在一起，而大型、高排量的车辆聚在另一组，这可能反映了经济型与高性能车型的区别。
* 列聚类（特征的聚类）：
列聚类展示了特征之间的相似性。例如，油耗（mpg）可能与重量和加速度有相关性，它们的聚类顺序上会显示出较短的分支，表示它们在一定程度上关联。
通过列聚类，可以观察到相关特征的聚集，帮助理解哪些特征可能对车辆性能或燃油效率有重要影响。
3. 聚类带来的数据洞察：
* 不同车款的分布：从聚类结果可以看到具有类似特征的车辆聚集在一起，可能揭示了数据集中一些规律。例如，基于 cylinders 和 horsepower 的相似性，某些高性能车款会分到一组，而经济型车款则在另一组。
* 特征相关性：列聚类的结果也可以帮助理解数据特征间的潜在关系。例如，重量和油耗可能较为接近，表明重量较大的车辆通常油耗较高，这符合车辆设计的常见现象。


4. 总结：
总体来说，这个热力图和聚类结果可以帮助我们直观地看到汽车数据集中的车型分布和特征关联，发现性能类似的车型类别，以及特征之间的相关性。

# 4. 绘制方法


热图的绘制通常使用Python的matplotlib库和seaborn库。这两者的heatmap函数都可以生成直观的热图，并提供丰富的参数来控制颜色、注释、数据范围等。以下是绘制热图的基本方法和参数介绍。


1. 使用 Matplotlib 绘制热图


matplotlib的plt.imshow()函数可以绘制热图，尤其适合展示二维数组数据。

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机数据
data = np.random.rand(10, 12)  # 10行12列的二维数组

# 绘制热图
plt.imshow(data, cmap='viridis', aspect='auto')  # 设置颜色映射和比例
plt.colorbar()  # 添加颜色条
plt.title("Heatmap with Matplotlib")
plt.show()
```

常用参数
* data：要绘制的二维数组或矩阵。
* cmap：指定颜色映射，如viridis、plasma、coolwarm等。
* aspect：控制图像宽高比，auto会根据数据自动调整，equal会固定每个单元格为正方形。
* colorbar()：显示颜色条，帮助解释颜色对应的数值范围。

2. 使用 Seaborn 绘制热图


seaborn的heatmap函数在显示数值数据时更为方便，支持注释、分层等功能。

```python
import seaborn as sns
import numpy as np

# 生成示例数据
data = np.random.rand(10, 12)  # 10行12列的随机数矩阵

# 绘制热图
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.title("Heatmap with Seaborn")
plt.show()
```

常用参数
* data：二维数组或DataFrame（pandas中的数据框），为要展示的数据。
* cmap：指定颜色映射，如YlGnBu、coolwarm、magma等。
* annot：布尔值，若设置为True，会在每个格子中显示数值。
* fmt：设置显示数值的格式，比如.1f表示保留一位小数。
* linewidths：控制每个单元格之间的边框宽度，有助于更好地分隔各格子。


其他常见参数
* vmin, vmax：指定颜色映射的数值范围，方便突出重要数据。
* cbar_kws：字典类型，可控制颜色条的属性（如{'orientation': 'horizontal'}使颜色条水平显示）。

3. 示例：带注释的相关性热图


可以使用Pandas生成数据的相关性矩阵，并绘制热图展示变量之间的相关性：

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 生成示例数据
data = pd.DataFrame(np.random.rand(100, 4), columns=["A", "B", "C", "D"])

# 计算相关性矩阵
corr = data.corr()

# 绘制相关性热图
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

常见参数：
* square：布尔值，使每个单元格为正方形。
* corr：相关性矩阵，通过data.corr()计算。
* annot=True：将每个相关系数显示在热图上，更清晰地展示变量关系。
